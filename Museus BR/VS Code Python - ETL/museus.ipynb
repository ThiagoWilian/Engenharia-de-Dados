Web Scraping

```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.options import Options 
from selenium.common.exceptions import NoSuchElementException

description = {}  # Aqui terá o nome e a descrição
not_found = [] # Aqui terá o nome das descrições não encontradas
```

```python
for pesquisa in null_description:
  
  texto = None
  config = Options()
  config.add_argument('--headless')
  navegador = webdriver.Chrome(options=config) 
  navegador.get('https://www.google.com.br/') 
  sleep(0.5)
  navegador.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(f"{pesquisa}")
  #sleep(0.5) >>>>>>>>>>  Aqui estava dando problema de click, devido ao sleep, que conforme a documentação deve ser evitada  <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
  
  botao = navegador.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[4]/center/input[1]').click()
                               
  sleep(0.5)
  try:
    if not texto:
      texto = str(navegador.find_element_by_xpath('//*[@id="kp-wp-tab-overview"]/div[1]/div/div/div/div/div/div[3]/div/div/div/div/span[1]').text)
      description[pesquisa] = texto
        
  except NoSuchElementException:
    pass

  try:
    if not texto:
      texto = str(navegador.find_element_by_xpath('/html/body/div[7]/div/div[11]/div[2]/div/div/div[2]/div/div[3]/div/div/div/div/div[1]/div/div/div/div/div/div[3]/div/div/div/div/span[1]').text)
      description[pesquisa] = texto
     
  except NoSuchElementException:
    pass

  try:
    if not texto:
      texto = str(navegador.find_element_by_xpath('/html/body/div[7]/div/div[11]/div[2]/div[1]/div/div[2]/div/div[5]/div/div/div/div/div[1]/div/div/div/div/div/div/div[1]/div/div/div/span[1]').text)              
      description[pesquisa] = texto
          
  except NoSuchElementException:
    pass

  try:
    if not texto:
      texto = str(navegador.find_element_by_xpath('/html/body/div[7]/div/div[11]/div[2]/div/div/div[2]/div/div[3]/div/div/div/div/div[1]/div/div/div/div/div/div[3]/div/div/div[1]/div/span[2]').text)              
      description[pesquisa] = texto
          
  except NoSuchElementException:
    pass

  try:
    if not texto:
      texto = str(navegador.find_element_by_xpath('/html/body/div[7]/div/div[11]/div[2]/div[1]/div/div[2]/div/div[4]/div/div/div/div/div[1]/div/div/div/div/div/div/div[1]/div/div/div/span[1]').text)      
      description[pesquisa] = texto
          
  except NoSuchElementException:
    pass

  try:
    if not texto:
      texto = str(navegador.find_element_by_xpath('/html/body/div[7]/div/div[11]/div[2]/div[1]/div/div[2]/div/div[5]/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div/div/span[1]').text)     
      description[pesquisa] = texto
          
  except NoSuchElementException:
    pass

  try:
    if not texto:
      texto = str(navegador.find_element_by_xpath('/html/body/div[7]/div/div[11]/div[2]/div/div/div[2]/div/div[4]/div/div/div/div/div/div/div/div/div/div/div/div/div/div/div/span[1]').text)             
      description[pesquisa] = texto
          
  except NoSuchElementException:
    pass

  if not texto:
    not_found.append(pesquisa) 

  navegador.close()
```

```python
# Adicionando as descrições adquiridas acima, no Dataframe
for i in description:
    linha = np.where(df_museu['name'] == i)
    df_museu['shortDescription'].loc[linha] = description[i]
```
